```python
import pandas as pd
```

```python
EVAL_DIR = "../evals"
DIRS = ["sta_lta_stw1_ltw2", 
 "sta_lta_stw1_ltw4", 
 "sta_lta_stw1.5_ltw3", 
 "sta_lta_stw1.5_ltw6", 
 "sta_lta_stw2_ltw4", 
 "sta_lta_stw2_ltw6", 
 "sta_lta_stw2_ltw8", 
 "sta_lta_stw2.5_ltw4", 
 "sta_lta_stw2.5_ltw5", 
 "sta_lta_stw2.5_ltw7.5", 
 "sta_lta_stw3_ltw4.5", 
 "sta_lta_stw3_ltw6",
 "sta_lta_stw3_ltw7.5",
 "sta_lta_stw3.5_ltw5",
 "sta_lta_stw3.5_ltw6.5",
 "sta_lta_stw3.5_ltw8",
 "sta_lta_stw4_ltw5",
 "sta_lta_stw4_ltw6.5",
 "sta_lta_stw4_ltw8",
 "sta_lta_stw4.5_ltw6",
 "sta_lta_stw4.5_ltw7",
 "sta_lta_stw4.5_ltw8"]

PARAMS = [{"stw":1, "ltw":2},
          {"stw":1, "ltw":4},
          {"stw":1.5, "ltw":3},
          {"stw":1.5, "ltw":6},
          {"stw":2, "ltw":4},
          {"stw":2, "ltw":6},
          {"stw":2, "ltw":8},
          {"stw":2.5, "ltw":4},
          {"stw":2.5, "ltw":5},
          {"stw":2.5, "ltw":7.5},
          {"stw":3, "ltw":4.5},
          {"stw":3, "ltw":6},
          {"stw":3, "ltw":7.5},
          {"stw":3.5, "ltw":5},
          {"stw":3.5, "ltw":6.5},
          {"stw":3.5, "ltw":8},
          {"stw":4, "ltw":5},
          {"stw":4, "ltw":6.5},
          {"stw":4, "ltw":8},
          {"stw":4.5, "ltw":6},
          {"stw":4.5, "ltw":7},
          {"stw":4.5, "ltw":8}]
```

```python
dfs = []
for i, dir in enumerate(DIRS):
    df = pd.read_csv("{}/{}/metrics.csv".format(EVAL_DIR, dir))
    df["short term window"] = PARAMS[i]["stw"]
    df["long term window"] = PARAMS[i]["ltw"]
    dfs.append(df)

df = pd.concat(dfs)
df1 = df.groupby(["short term window", "long term window", "testing"]).mean().reset_index()
df2 = df.groupby(["short term window", "long term window", "testing"]).std().reset_index()
df1["auc"] = df1["auc"].apply(lambda x: "${:.5f}".format(x)) + " \pm " + df2["auc"].apply(lambda x: "{:.5f}$".format(x))
df1.drop(columns=["pre", "rec", "f1_score", "av_pre", "split", "epoch"], inplace=True)

df_instance = df1[df1.testing == "instance"]
df_stead = df1[df1.testing == "stead"]

df_instance.drop(columns=["testing"], inplace=True)
df_stead.drop(columns=["testing"], inplace=True)

df_instance.rename(columns={"auc":"INSTANCE AUC"}, inplace=True)
df_stead.rename(columns={"auc":"STEAD AUC"}, inplace=True)

df_instance.sort_values(by=["short term window", "long term window"], inplace=True)
df_stead.sort_values(by=["short term window", "long term window"], inplace=True)

df = pd.merge(df_instance, df_stead, on=["short term window", "long term window"])
print(df.to_latex(index=False, escape=False, float_format="%.5f"))
```
